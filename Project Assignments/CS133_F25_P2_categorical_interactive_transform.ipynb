```python
# Your code here . . .
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset_file_path = '/content/drive/MyDrive/Nutrition__Physical_Activity__and_Obesity_-_Behavioral_Risk_Factor_Surveillance_System.csv'
```

```python
df = pd.read_csv(dataset_file_path)
df.info()
```

```python
df.head(3)
```

```python
df_obesity = df.loc[df['Class'].str.contains('Obesity', na = False)]
df_obesity.rename(columns={'LocationDesc':'Location'}, inplace=True)
display(df_obesity)
```

```python
sns.countplot(data = df_obesity,
            x = 'Location'
           );
plt.xticks(rotation=90)
plt.title('Obesity Case count per Location (State)')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_obesity = df.dropna(subset=["Sex", "Data_Value", "YearStart"])

df_obesity["Sex"] = df_obesity["Sex"].replace({
    "Male": "Male",
    "Female": "Female",
    "Overall": "Overall"
})

df_obesity["YearStart"] = df_obesity["YearStart"].astype(str)

sns.catplot(
    data=df_obesity,
    x="Sex",
    y="Data_Value",
    kind="strip",
    col="YearStart",
    jitter=True,
    alpha=0.5,
    height=5,
    aspect=1
)

plt.subplots_adjust(top=0.85)
plt.suptitle("Data Values by Sex, Split by Year")

plt.show()
```

### Q4: Create an interactive plot
Create an interactive plot using plotly. Your plot can be a previously asked question or a new question, however explain how this interaction strengthen what you want to communicate with your audience.

This interactive bar chart displays the average obesity percentage for each U.S. state. Each bar represents one state, and users can hover over a bar to see the exact obesity percentage. The interactivity helps to easily identify which states have higher or lower obesity rates, such as Mississippi and West Virginia showing higher averages. This makes the data more engaging and allows for quick comparisons between states.

```python
# Your code here . . .
import plotly.express as px
import plotly as py
import plotly.graph_objects as go
```

```python
# Calculate the average Data_Value for each state in the df_obesity DataFrame
average_obesity_per_state = df_obesity.groupby('Location')['Data_Value'].mean().reset_index()

fig = px.bar(average_obesity_per_state,
             x='Location',
             y='Data_Value',
             title='Average Obesity Percentage by State',
             labels={'Location': 'State', 'Data_Value': 'Average Obesity Percentage'})
fig.update_layout(xaxis_tickangle=-90)
fig.show()
```
